### Concinating DataFrame

```python
class_a=pd.DataFrame({'Mat':[45,68,35,65],
                     'Eng':[54,87,36,52],
                     'Sci':[66,77,88,70],
                     'Chem':[54,79,58,76]},
                    index=['Ravi','Shyam','Saurav','Abhi'])
class_a
```

```python
class_b=pd.DataFrame({'Mat':[50,87,69,71],
                     'Eng':[66,50,87,69],
                     'Sci':[34,66,69,71],
                     'Chem':[54,87,34,66]},
                    index=['Shaan','Mahi','Viru','Mona'])
class_b
```

```python
class_c=pd.DataFrame({'Mat':[66,50,87,56],
                     'Eng':[56,87,89,71],
                     'Sci':[77,88,70,52],
                     'Chem':[34,56,56,87]},
                    index=['Astha','Rohit','Rahul','Shital'])
class_c
```

```python
# row wize concatination 

all_students = pd.concat([class_a , class_b , class_c] , axis = 0 )
all_students
```

```python
# column wise concatination 
all_students = pd.concat([class_a , class_b , class_c] , axis = 1 )
all_students
```

```python
all_students_nullless = all_students.fillna(value = 'unattempted')
all_students_nullless
```
